```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('Motor_Vehicle_Registrations_Dashboard_data.csv')
df = df[df["state"] != "Puerto Rico"]
df['state'] = df['state'].str.replace('\(2\)', '', regex=True)
df
```

```python
# @title year vs Bus

def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['Bus']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('year', ascending=True)
df_averages = df_sorted.groupby('year', as_index=False)['Bus'].mean()
_plot_series(df_averages, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('Bus')
```
