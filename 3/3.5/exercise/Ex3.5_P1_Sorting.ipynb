```python
import pandas as pd
```

```python
df = pd.read_excel('Global Superstore 2016.xlsx', sheet_name='Orders')
print(df.shape)
df.head()
```

```python
print("3.5 Part 1")
selected_states = ['North Carolina', 'Georgia', 'Virginia']

df_Q1 = df.sort_values(by=['Segment', 'Customer Name', 'Sales'], ascending=[True, True, False])
df_Q1 = df.query('State in @selected_states and Sales > 3000')
print(df_Q1.shape)
df_Q1
```

```python
print("3.5 Part 2")
df = pd.read_csv('Cars.csv')

df_Q2 = df.sort_values(by=['MPG_Hwy', 'Vehicle_Model'], ascending=[False, True], inplace=True)
df_Q2 = df.query('Vehicle_Make == "Chevrolet"')
df_Q2 = df_Q2.head(5)
print(df_Q2.shape)
df_Q2
```

```python
print("3.5 Part 3: In 2021, what were the top 10 leading causes of death in the US?")

df_Q3 = pd.read_csv('Leading_Causes_of_Death_2021.csv')
df_Q3 = df_Q3.sort_values(by=['Deaths'], ascending=False)
df_Q3 = df_Q3.head(10)
print(df_Q3.shape)
df_Q3
```
