# Ex: 4.1: Plotting with Seaborn

This notebook demonstrates how to use the Seaborn plotting package to visualize data.  The built-in Seaborn dataset, ```mpg```, will be used to demonstrate the plotting capabilities of Seaborn. 

- **Demonstrates**: 
    - How to create a scatter plot with Seaborn

---

### Step 1.

* Run the following code block below to load the "mpg" dataset from sns and assign it to the variable ```mpg```.  This loads the data into a pandas dataframe.  Take a peek at the data using the ```head()``` method. 

````python
import seaborn as sns
import matplotlib.pyplot as plt
mpg = sns.load_dataset("mpg")
mpg['make'] = mpg['name'].apply(lambda s: s.split()[0])
mpg.head()
````

### Step 2.
* Use ```sns.relplot``` to make a scatterplot of the "horsepower" (on the x-axis)  vs. "weight" (on the y-axis) columns of the mpg dataframe. To do this you will need to specify the arguments ```x``` and ```y``` with the strings that correspond to the variable names and the argument ```data``` with the source data frame. Include the ```hue``` option argument to color the points by the ```model_year``` column.

````python
sns.relplot(data=mpg, x='horsepower', y='weight', hue='model_year')
````
