# Part 1

```python
from bokeh.layouts import column
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.layouts import row
import yfinance as yf
import pandas as pd

df_amzn = yf.download(tickers = "AMZN", start = "2024-01-01", end = "2025-01-01", multi_level_index = False).reset_index()
df_goog = yf.download(tickers = "GOOG", start = "2024-01-01", end = "2025-01-01", multi_level_index = False).reset_index()
df_abnb = yf.download(tickers = "ABNB", start = "2024-01-01", end = "2025-01-01", multi_level_index = False).reset_index()
print(df_amzn.dtypes, df_goog.dtypes, df_abnb.dtypes)
output_notebook()
data_source_amzn = ColumnDataSource(df_amzn)
data_source_goog = ColumnDataSource(df_goog)
data_source_abnb = ColumnDataSource(df_abnb)
hover_high = HoverTool(tooltips=[("High Stock Price", "@High")])
# Instantiate the figure object for 'High' and 'Low'
fig_high_amzn = figure(title='Amazon Stock Prices', x_axis_type='datetime', tools=[hover_high], width=400, height=400)
fig_high_goog = figure(title='Google Stock Prices', x_axis_type='datetime', tools=[hover_high], width=400, height=400)
fig_high_abnb = figure(title='Airbnb Stock Prices', x_axis_type='datetime', tools=[hover_high], width=400, height=400)
fig_high_combined = figure(title='Combined Stock Prices', x_axis_type='datetime', tools=[hover_high], width=1200, height=400)

# Add the 'High' Stock Price line to the first figure
fig_high_amzn.line(x='Date', y='High', color='blue', source=data_source_amzn, legend_label='AMZN')
fig_high_goog.line(x='Date',  y='High', color='red', source=data_source_goog, legend_label='GOOG')
fig_high_abnb.line(x='Date',  y='High', color='green', source=data_source_abnb, legend_label='ABNB')
fig_high_combined.line(x='Date',  y='High', color='green', source=data_source_amzn, legend_label='AMZN')
fig_high_combined.line(x='Date', y='High', color='red', source=data_source_goog, legend_label='GOOG')
fig_high_combined.line(x='Date', y='High', color='blue', source=data_source_abnb, legend_label='ABNB')

# Use row layout to place figures side by side
final_figure_row = row(fig_high_amzn, fig_high_goog, fig_high_abnb)
final_figure_column = column(fig_high_combined)

# Display the layout
show(final_figure_row)
show(final_figure_column)
```

# Part 2

```python
df = pd.read_csv('top-movies.csv')
df.sort_values('year', ascending=True, inplace=True)
print(df.shape)
df.head
output_notebook()
movies = ColumnDataSource(df)
hover_card = HoverTool(tooltips=[("Title", "@title"), ("Runtime", "@runtime"), ("Year", "@year")])

# Instantiate the figure object
fig = figure(title='Runtime of Movies Over Different Years', x_axis_label='Year', y_axis_label='Runtime (in minutes)', tools=[hover_card])

# Call the line function
fig.line(x='year', y='runtime', source=movies, line_width=2, color='orange', line_dash='solid')
fig.scatter(x='year', y='runtime', source=movies, size=8, color='purple')

# display the plot in the notebook
show(fig)
```
