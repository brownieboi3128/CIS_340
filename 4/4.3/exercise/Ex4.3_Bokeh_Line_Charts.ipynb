```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
import pandas as pd
df = pd.read_csv('avocados.csv')
print(df.shape)
df.head(2)
df.dtypes
output_notebook()
source = ColumnDataSource(df)
```

# Part 1 
Using the avocados.csv data file, create the below Bokeh line chart displaying the average annual avocado prices from California.

```python
# Instantiate the figure object
fig = figure(title='Average Annual Avocado Price by Region', x_axis_label='Year', y_axis_label='Prices', background_fill_color='rgb(89, 44, 136)')

# Call the line function
fig.line(x='Year', y='California', source=source, line_width=7, color='orange', line_dash='dotted', legend_label='California')

# display the plot in the notebook
show(fig)
```

# Part 2
create the below Bokeh multi-line chart displaying the average annual avocado prices from Mexico, California, and Peru.

```python
# Instantiate the figure object
fig = figure(title='Average Annual Avocado Price by Region', x_axis_label='Year', y_axis_label='Prices', background_fill_color='rgb(89, 44, 136)')

# Call the line function
fig.line(x='Year', y='California', source=source, line_width=7, color='orange', line_dash='dotted', legend_label='California')
fig.line(x='Year', y='Mexico', source=source, line_width=7, color='green', line_dash='solid', legend_label='Mexico')
fig.line(x='Year', y='Peru', source=source, line_width=7, color='red', line_dash='dashed', legend_label='Peru')


# display the plot in the notebook
show(fig)
```

# Part 3
Use the top-movies.csv data file and sort the dataframe by year. Then, create a simple line chart showing the runtime of movies over the different years.

```python
df = pd.read_csv('top-movies.csv')
df.sort_values('year', ascending=True, inplace=True)
print(df.shape)
df.head
output_notebook()
movies = ColumnDataSource(df)
# Instantiate the figure object
fig = figure(title='Runtime of Movies Over Different Years', x_axis_label='Year', y_axis_label='Runtime (in minutes)')

# Call the line function
fig.line(x='year', y='runtime', source=movies, line_width=2, color='orange', line_dash='solid')

# display the plot in the notebook
show(fig)
```
